# Key Stats - Total Owners

Counts are from table "invocations"

```python
%%pyspark
df = spark.sql("""
    SELECT
        HashOwner, COUNT(HashOwner) as Counts
    FROM
        invocations
    GROUP BY HashOwner
    ORDER BY Counts DESC

""") 
df.summary().show()
```

# Key Stats - Top 10 Owners

```python
top10_owners_df = df.limit(10)
top10_owners_df.write.mode("overwrite").saveAsTable("top10_owners")
display(top10_owners_df)
```

# Preparing Sub_Datasets

```python
%%pyspark
df = spark.sql("""
    SELECT
        *
    FROM
        app_memory NATURAL JOIN top10_owners
""") 
df.write.mode("overwrite").saveAsTable("app_memory_bytopOwners")
display(df.limit(10))
```

```python
%%pyspark
df = spark.sql("""
    SELECT
        *
    FROM
        function_durations NATURAL JOIN top10_owners
""") 
df.write.mode("overwrite").saveAsTable("function_durations_bytopOwners")
display(df.limit(10))
```

```python
%%pyspark
df = spark.sql("""
    SELECT
        *
    FROM
        invocations NATURAL JOIN top10_owners
""") 
df.write.mode("overwrite").saveAsTable("invocations_bytopOwners")
display(df.limit(10))
```

```python
%%pyspark
df = spark.sql("""
    SELECT
        *
    FROM
        app_memory NATURAL JOIN top10_apps
""") 
df.write.mode("overwrite").saveAsTable("app_memory_bytopApps")
df = spark.sql("""
    SELECT
        *
    FROM
        function_durations NATURAL JOIN top10_apps
""") 
df.write.mode("overwrite").saveAsTable("function_durations_bytopApps")
df = spark.sql("""
    SELECT
        *
    FROM
        invocations NATURAL JOIN top10_apps
""") 
df.write.mode("overwrite").saveAsTable("invocations_bytopApps")
```

```python
%%pyspark
df = spark.sql("""
    SELECT
        *
    FROM
        app_memory NATURAL JOIN top10_functions
""") 
df.write.mode("overwrite").saveAsTable("app_memory_bytopFunctions")
df = spark.sql("""
    SELECT
        *
    FROM
        function_durations NATURAL JOIN top10_functions
""") 
df.write.mode("overwrite").saveAsTable("function_durations_bytopFunctions")
df = spark.sql("""
    SELECT
        *
    FROM
        invocations NATURAL JOIN top10_functions
""") 
df.write.mode("overwrite").saveAsTable("invocations_bytopFunctions")
```

# Export Sub_DataSet - Top 10 Users

Join all original sub-datasets and export all columns relating to the Top 10 users

```python
%%pyspark
df = spark.sql("""
    WITH slim_appmem_invoke
    AS
    (
        SELECT
            *
        FROM
            app_memory_bytopOwners NATURAL JOIN function_durations_bytopOwners
    ),
    slim_function_appmem
    AS
    (
        SELECT
            *
        FROM
            function_durations_bytopOwners NATURAL JOIN invocations_bytopOwners
    )
    SELECT *
    FROM slim_appmem_invoke NATURAL JOIN slim_function_appmem
""") 
#df.summary().show()
df.write.option("header","TRUE").csv("abfss://genfilesys@gensynapse.dfs.core.windows.net/Azure-functionDataset-2019/synapse-data/top10_USERS")
display(df.limit(10))
```
